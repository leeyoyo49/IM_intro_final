```python
from Restaurant.restaurant import *

Item_Queue = ItemQueue()

# Don't need to care about this part
Test_order_id = [1,6,10,12,13,14,1,15,18,19,8,9,5,6,18,19]

# when creating an new order, just put a list filled with item id eg.Test_order_id = [1,6,10,12,13,14,1,15,18,19,8,9,5,6,18,19]
Test_order = Order(Test_order_id)

# after creating, add the order to the queue
Item_Queue.add_order(Test_order)

# Then you need to check Item_Queue.all_waiting_arranged to see the order of the items
# Item_Queue.all_waiting_arranged looks like this, there's 7 list in it, each list are sorted according to its priority.
# [[item1, item2], 
#  [item3, item4], 
#  [item5, item6], 
#  [item7, item8], 
#  [item9, item10], 
#  [item11, item12], 
#  [item13, item14]]
print("\n\n")
print(Item_Queue.all_waiting_arranged)

# After that, you can pop the item from waiting to main
# Item_Queue.pop_wait(0) will pop the first item in station 0 to main
# Item_Queue.pop_wait will return the item that is popped, so you can use it to check the item
print(Item_Queue.pop_wait(0).name)
print(Item_Queue.all_waiting_arranged)
print(Item_Queue.all_main)

# After the item is done, you can pop the item from main
# Item_Queue.pop_main(0) will pop the first item in station 0 to main, it also return the item that is popped
# You'll need to check the item popped's Item_completed to see if the item is completed, since an item might have 2 stages
# You then need to update the order list you created
print(Item_Queue.pop_main(0).Item_completed)
print(Item_Queue.all_waiting_arranged)
print(Item_Queue.all_main)
```
